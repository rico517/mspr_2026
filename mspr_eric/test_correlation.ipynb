```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./1er_tour_paris_2022.csv')
```

```python
df=df.drop(['Libelle de la circonscription','Code du departement','Libelle du departement','Code de la circonscription','Nom','Nom2','Nom3','Nom4','Nom5','Nom6','Nom7','Nom8','Nom9','Nom10','Nom11','Nom12'],axis=1)
```

```python
corr=df.corr()
```

```python
# 3. Configurer l'esthétique de la figure
plt.figure(figsize=(10, 8))
sns.set_theme(style="white")

# 4. Créer la heatmap
# annot=True affiche les coefficients dans les cases
# cmap définit la palette de couleurs (coolwarm est idéal pour la corrélation)
# fmt=".2f" arrondit à deux décimales
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)

# 5. Ajouter un titre et afficher
plt.title('Matrice de Corrélation - Dataset Iris')
plt.show()
```
